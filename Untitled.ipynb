```python
# Import pandas
import pandas as pd
from bokeh.plotting import figure,ColumnDataSource
from bokeh.io import output_notebook, show
from bokeh.models import HoverTool, NumeralTickFormatter
from bokeh.layouts import gridplot
from bokeh.charts import Histogram, Bar, output_file, Scatter,BoxPlot, output_file, show


# Assign spreadsheet filename: file
file = 'imd_student_blind.xlsx'

# Load spreadsheet: xl
xl = pd.ExcelFile(file)

# Print sheet names
print(xl.sheet_names)
```

```python
# Load a sheet into a DataFrame by index: df
data = xl.parse(0)
TOOLS = 'box_zoom,box_select,crosshair,resize,reset,wheel_zoom, pan'

# Print the head of the DataFrame df
data.head()
```

```python
source = ColumnDataSource(data)
data['count'] = 0
data.head()
```

```python
d = data[data["status.disciplina"] == "Reprovado"]
df1 = pd.DataFrame(d.groupby(['disciplina_ID','ano_disciplina'])['count'].count()).reset_index()
df1.head(10)
```

```python
dataYear2014_ID = df1[df1["ano_disciplina"] == 2014]["disciplina_ID"]
dataYear2014_count = df1[df1["ano_disciplina"] == 2014]["count"]


dataYear2015_ID = df1[df1["ano_disciplina"] == 2015]["disciplina_ID"]
dataYear2015_count = df1[df1["ano_disciplina"] == 2015]["count"]

dataYear2016_ID = df1[df1["ano_disciplina"] == 2016]["disciplina_ID"]
dataYear2016_count = df1[df1["ano_disciplina"] == 2016]["count"]

dataYear2016_count.head(100)
```

```python

#source = ColumnDataSource(df1)

Reprovados_Disc2014 = figure(title='Scatter Nº de reprovados pelas disciplinas, Ano de 2014', x_axis_label='Disciplina', y_axis_label='Nº Reprovados',tools=[TOOLS])
Reprovados_Disc2015 = figure(title='Scatter Nº de reprovados pelas disciplinas, Ano de 2015', x_axis_label='Disciplina', y_axis_label='Nº Reprovados',tools=[TOOLS])
Reprovados_Disc2016 = figure(title='Scatter Nº de reprovados pelas disciplinas, Ano de 2016', x_axis_label='Disciplina', y_axis_label='Nº Reprovados',tools=[TOOLS])


Reprovados_Disc2014.circle(dataYear2014_ID,dataYear2014_count,size=10,color = 'green')
Reprovados_Disc2015.circle(dataYear2015_ID,dataYear2015_count,size=10,color = 'navy')
Reprovados_Disc2016.circle(dataYear2016_ID,dataYear2016_count,size=10,color = 'red')

```

## Gráfico 1 - Número de Reprovados x Disciplina em cada ano.

Nesta seção utilizamos os dados do número de reprovados em cada uma das disciplinas disponíveis, pelos anos em que elas foram ofertadas.

"Há um sistema pré-estabelecido ao qual o aluno precisa se adaptar, se inserir, como um ritual de passagem. Quem não for capaz de cumprir etapas estabelecidas pela coordenação da disciplina, provavelmente, será reprovado.”

#### Fabiana Colombo Garzella, 
**A disciplina de Cálculo I: a análise das relações entre as práticas pedagógicas do professor e seus impactos nos alunos**

O Número de reprovados em algumas disciplinas do Curso do Bacharelado em Tecnologia da Informação, tem crescido rapidamente e é algo que deve ser revisado. Pensando nisso resolvemos analisar o número de reprovados em cada uma das disciplinas disponíveis, pelos anos em que elas foram ofertadas.

Como visto, algumas disciplinas, a cada ano, tem seu número de reprovações aumentando. Além de outros possíveis fatores, a forma como a disciplina está sendo ofertada, a relação do aluno com os professores ou a quantidade de alunos em uma única turma, deve ser revisada pela coordenação ou setor pedagógico.
Nos gráficos, podemos destacar a disciplina 5, que a cada ano aumenta o número de reprovados, chegando a 2016 com mais de 250 reprovações naquele ano. A disciplina 2, apesar de ter mantido esse valor de 2014 para 2015 com cerca de 50, teve em torno de 230 reprovações em 2016. A disciplina 3 também teve um aumento considerável, já que em 2014 o valor era de aproximadamente 50 para em torno de 180 em 2016.

```python
row1 = [Reprovados_Disc2014,Reprovados_Disc2015]

# Create a list containing plots p3 and p4: row2
row2 = [Reprovados_Disc2016]

layout = gridplot([row1,row2],plot_width=450, plot_height=450)

output_notebook()

# Display the plot
show(layout)
```
