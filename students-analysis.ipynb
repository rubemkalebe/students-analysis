## Análise 8 - Evolução temporal das disciplinas

Nesta seção utilizamos a quantidade de aprovados e reprovados, além do total de matriculados em cada disciplina, para mostrar o desenvolvimento de cada disciplina no decorrer dos anos em que foram ofertadas.

A análise do quanto uma disciplina vem aprovando, ou reprovando, é útil para encontrar disciplinas nas quais os estudantes apresentam dificuldades ou ainda, disciplinas que eventualmente têm se tornado mais complicada para eles. Um gráfico de linhas é ideal para apresentar essa evolução no decorrer do tempo.

À primeira vista, percebe-se que os alunos apresentam mais dificuldades nas disciplinas 3, 5 e 6. A disciplina 0 também apresenta-se de complexidade razoável e a disciplina 2, a qual antes não poderia ser considerada de grande complexidade, passou a reprovar a maior parte dos seus alunos.

As disciplinas 1 e 4 aparentam serem melhor aproveitadas pelos alunos e a tendência de crescimento na quantidade de matriculados mostra que essas disciplinas não sofreram mudanças que alterassem a tendência de aprovação.

A disciplina 6 apresenta uma tendência preocupante de "reprovar mais do que aprovar". Com o passar dos anos, a quantidade de aprovados praticamente se manteve constante, enquanto que a quantidade de reprovados cresceu linearmente.

```python
df1 = data[data['ano_disciplina'] != 2017]
df1['count_reprovados'] = 0
df1['count_aprovados'] = 0
df1['count_total'] = 0
df1['count_rep_periodo'] = 0

reprovados = df1[df1["status.disciplina"] == "Reprovado"]
aprovados = df1[df1["status.disciplina"] == "Aprovado"]

reprovados_count = pd.DataFrame(reprovados.groupby(['disciplina_ID','ano_disciplina'])['count_reprovados'].count()).reset_index()
aprovados_count = pd.DataFrame(aprovados.groupby(['disciplina_ID','ano_disciplina'])['count_aprovados'].count()).reset_index()
total_count = pd.DataFrame(df1.groupby(['disciplina_ID','ano_disciplina'])['count_total'].count()).reset_index()
rep_periodo_count = pd.DataFrame(reprovados.groupby(['ano_disciplina', 'periodo_disciplina'])['count_rep_periodo'].count()).reset_index()
```

```python
from bokeh.models.widgets import Panel, Tabs
from bokeh.io import output_file, output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool

def getSource(disc):
    return ColumnDataSource(data = {
        'anos': [str(x) for x in reprovados_count[reprovados_count["disciplina_ID"] == disc]['ano_disciplina']],
        'reprovados': reprovados_count[reprovados_count["disciplina_ID"] == disc]['count_reprovados'],
        'aprovados' : aprovados_count[aprovados_count["disciplina_ID"] == disc]['count_aprovados'],
        'total' : total_count[total_count["disciplina_ID"] == disc]['count_total']
    }
)

def getFigure(disc):
    source = getSource(disc)
    p = figure(plot_width = 600, plot_height = 400, x_axis_label = 'Anos', y_axis_label = 'Número de alunos',
               x_range = source.data['anos'], title = 'Aprovados/Reprovados na Disciplina ' + str(disc))
    p.line(x = 'anos', y = 'reprovados', color = "red", line_width = 3, source = source, legend='Reprovados')
    p.circle(x = 'anos', y = 'reprovados', fill_color = "white", size = 7, source = source)
    p.line(x = 'anos', y = 'aprovados', color = "green",line_width = 3, source = source, legend='Aprovados')
    p.circle(x = 'anos', y = 'aprovados', fill_color = "white", size = 7, source=source)
    p.line(x = 'anos', y = 'total', color = "blue", line_width = 3, source = source, legend='Total')
    p.circle(x = 'anos', y = 'total', fill_color = "white", size = 7, source = source)
    p.legend.location = "top_left"
    return p

tab1 = Panel(child = getFigure(0), title = "Disciplina 0")
tab2 = Panel(child = getFigure(1), title = "Disciplina 1")
tab3 = Panel(child = getFigure(2), title = "Disciplina 2")
tab4 = Panel(child = getFigure(3), title = "Disciplina 3")
tab5 = Panel(child = getFigure(4), title = "Disciplina 4")
tab6 = Panel(child = getFigure(5), title = "Disciplina 5")
tab7 = Panel(child = getFigure(6), title = "Disciplina 6")
tabs = Tabs(tabs = [tab1, tab2, tab3, tab4, tab5, tab6, tab7])


output_notebook()
#output_file("results/analise8.html")
show(tabs)
```

## Análise 9 - Evolução temporal dos períodos

Nesta seção utilizamos a quantidade de reprovados para mostrar o desenvolvimento dos períodos no decorrer dos anos para determinar se os períodos estão equilibrados.

Entre 2014 e 2015 os períodos seguiam a mesma tendência e o 2º Período se apresentava como mais desafiador aos alunos. Todavia, em 2016 o 1º Período teve um grande salto no número de reprovados (reprovou mais de 400 alunos a mais do que no ano anterior) e passou a ser, momentaneamente, o semestre mais desafiador. Essa tendência de um ano de 2016 mais complicado para os alunos já foi notada em análises anteriores que apresentaram altos níveis de reprovação nas disciplinas 2, 3, 5 e 6.

O 2º Período manteve sua tendência.

```python
from bokeh.models.widgets import Panel, Tabs
from bokeh.io import output_file, output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool

source = ColumnDataSource(data = {
        'anos': [str(x) for x in rep_periodo_count[rep_periodo_count["periodo_disciplina"] == 1]['ano_disciplina']],
        'reprovados1': rep_periodo_count[rep_periodo_count["periodo_disciplina"] == 1]['count_rep_periodo']
    }
)

source2 = ColumnDataSource(data = {
        'anos': [str(x) for x in rep_periodo_count[rep_periodo_count["periodo_disciplina"] == 2]['ano_disciplina']],
        'reprovados2' : rep_periodo_count[rep_periodo_count["periodo_disciplina"] == 2]['count_rep_periodo']
    }
)

p = figure(plot_width = 600, plot_height = 400, x_axis_label = 'Anos', y_axis_label = 'Número de alunos',
           x_range = source.data['anos'], title = 'Reprovados por período')
p.line(x = 'anos', y = 'reprovados1', color = "purple", line_width = 3, source = source, legend='1º Período')
p.circle(x = 'anos', y = 'reprovados1', fill_color = "white", size = 7, source = source)
p.line(x = 'anos', y = 'reprovados2', color = "orange",line_width = 3, source = source2, legend='2º Período')
p.circle(x = 'anos', y = 'reprovados2', fill_color = "white", size = 7, source=source2)
p.legend.location = "top_left"

output_notebook()
#output_file("results/analise9.html")
show(p)
```

## Análise 10 - Evolução temporal dos alunos que cancelaram ou trancaram

Nesta seção analisamos a quantidade de alunos que cancelaram ou trancaram no decorrer dos anos, para tentar fazer um paralelo com análises anteriores, principalmente as que envolvem número de reprovações.

Percebe-se que o número de alunos que cancelaram ou trancaram cresceu além da tendência dos anos anteriores. Isso deve ter sido incentivado pelo baixo desempenho dos alunos naquele ano.

```python
df2 = data[data['ano_disciplina'] != 2017]
df2['count_cancelado'] = 0
df2['count_trancado'] = 0

cancelado = df2[df2["status"] == "CANCELADO"]
trancado = df2[df2["status"] == "TRANCADO"]

cancelado_count = pd.DataFrame(cancelado.groupby(['a_ID','ano_disciplina'])['count_cancelado'].count()).reset_index()
trancado_count = pd.DataFrame(trancado.groupby(['a_ID','ano_disciplina'])['count_trancado'].count()).reset_index()

cancelado_dic = {}
trancado_dic = {}

for ind in set(cancelado_count['a_ID']):
    a = cancelado_count[cancelado_count['a_ID'] == ind]['ano_disciplina'].tail(1)
    if str(int(a)) in cancelado_dic:
        cancelado_dic[str(int(a))] += 1
    else:
        cancelado_dic[str(int(a))] = 1
        
for ind in set(trancado_count['a_ID']):
    a = trancado_count[trancado_count['a_ID'] == ind]['ano_disciplina'].tail(1)
    if str(int(a)) in trancado_dic:
        trancado_dic[str(int(a))] += 1
    else:
        trancado_dic[str(int(a))] = 1
```

```python
from bokeh.models.widgets import Panel, Tabs
from bokeh.io import output_file, output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Spectral6

source1 = ColumnDataSource(data = {
        'anos': [str(x) for x in sorted(cancelado_dic.keys())],
        'cancelado': [cancelado_dic[x] for x in sorted(cancelado_dic.keys())]
    }
)

source2 = ColumnDataSource(data = {
        'anos': [str(x) for x in sorted(trancado_dic.keys())],
        'trancado': [trancado_dic[x] for x in sorted(trancado_dic.keys())]
    }
)


p = figure(plot_width = 600, plot_height = 400, x_axis_label = 'Anos', y_axis_label = 'Número de alunos',
           x_range = source.data['anos'], title = 'Situação dos alunos')

p.line(x = 'anos', y = 'cancelado', color = Spectral6[0], line_width = 3, source = source1, legend = 'Cancelado')
p.circle(x = 'anos', y = 'cancelado', fill_color = "white", size = 7, source = source1)

p.line(x = 'anos', y = 'trancado', color = Spectral6[1], line_width = 3, source = source2, legend='Trancado')
p.circle(x = 'anos', y = 'trancado', fill_color = "white", size = 7, source=source2)
p.legend.location = "top_left"

output_notebook()
#output_file("results/analise10.html")
show(p)
```
